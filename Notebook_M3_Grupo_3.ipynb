```python
# Retira os espaços e substitui "PessoaColaboradora" por "PC"
dtg = df.select_dtypes(include = "object").columns
for a in dtg:
  df[a] = df[a].str.replace(' ', '')
  df[a] = df[a].str.replace('PessoaColaboradora', 'PC')

# Criação de colunas Mulher e Homem 
df['Mulher'] = np.nan
df['Homem'] = np.nan

# Analisa se o funcionário é homem ou mulher e registra com 1(true) ou 0(false) na respectiva coluna
for i in df.index:
  if df['Genero'][i] == 'Feminino':
    df['Mulher'][i] = 1
    df['Homem'][i] = 0
  elif df['Genero'][i] == 'Masculino':
    df['Mulher'][i] = 0
    df['Homem'][i] = 1
```

```python
df
```

```python
# Criação de colunas Solteiro, Casado/UniãoEstável e Divorciado/Separado
df['Solteiro'] = np.nan
df['Casado/UniãoEstável'] = np.nan
df['Divorciado/Separado'] = np.nan

# Analisa o estado civil do funcionário e registra com 1(true) ou 0(false) na respectiva coluna
for i in df.index:
  if df['Estado Civil'][i] == 'Solteiro':
    df['Solteiro'][i] = 1
    df['Casado/UniãoEstável'][i] = 0
    df['Divorciado/Separado'][i] = 0
  elif df['Estado Civil'][i] == 'Casado' or df['Estado Civil'][i] == 'UniãoEstável':
    df['Solteiro'][i] = 0
    df['Casado/UniãoEstável'][i] = 1
    df['Divorciado/Separado'][i] = 0
  elif df['Estado Civil'][i] == 'Divorciado' or df['Estado Civil'][i] == 'Separado':
    df['Solteiro'][i] = 0
    df['Casado/UniãoEstável'][i] = 0
    df['Divorciado/Separado'][i] = 1
```
