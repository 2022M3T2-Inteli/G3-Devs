```python
#Código para chamar a tabela Reconhecimento

df_reconhecimento = pd.read_excel('/content/drive/Shareddrives/modoulo3-grupo3/Base Colaboradores Everymind_Inteli_2020 a 2022vModelo Preditivo.xlsx', 'Reconhecimento')

df_reconhecimento
```

```python
#código para converter dentro da coluna "Situação" da tabela Reconhecimento os dados de "Afastado" para "Ativo"

ativo_reconhecimento = 'Ativo'

def situacao_afastado (col):
    return (col.str.replace('Afastado','Ativo'))
    
df_reconhecimento['Situação'] = situacao_afastado (df_reconhecimento.Situação)

df_reconhecimento
```

```python
#Código para apagar da tabela Reconhecimento algumas colunas

df_reconhecimento = df_reconhecimento.drop('Matricula',  axis=1)
df_reconhecimento = df_reconhecimento.drop('Alterou Função',  axis=1)
df_reconhecimento = df_reconhecimento.drop('Motivo',  axis=1)
df_reconhecimento = df_reconhecimento.drop('Novo Cargo',  axis=1)


df_reconhecimento
```

```python
table = pd.pivot_table(df_reconhecimento, values='Data de Admissão', index=['Codinome', 'Situação'],
    columns=['Data Vigência'], aggfunc=np.sum)
table
```
